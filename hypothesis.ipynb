# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
hourly_salaries = salaries[salaries['Salary or Hourly'] == 'Hourly']
hourly_salaries
```

```python
# Your code here:
#H0: the hourly wage of all hourly workers is equal = $30/h
#H1: the hourly wage of all hourly workers is (significantly different) to <> $30/h


from scipy.stats import ttest_1samp
import scipy.stats
stat, pval = ttest_1samp( hourly_salaries['Hourly Rate'], 30)
```

```python
print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval)
```

We reject the hypothesis, given that the p value is much lower than the 0.05 expected error having 95% confidence interval

We are also curious about salaries in the police force. The chief of police in Chicago claimed in a press briefing that salaries this year are higher than last year's mean of $86000/year a year for all salaried employees. Test this one sided hypothesis using a 95% confidence interval.

Hint: A one tailed test has a p-value that is half of the two tailed p-value. If our Null hypothesis is less than or equal to, then to reject, the test statistic must be positive (rejects the <) and the p-value below 0.05 (rejects the =).

```python
Annual_salaries = salaries[salaries['Salary or Hourly'] == 'Salary'] 
Annual_salaries
```

```python
police_salaries= Annual_salaries[Annual_salaries['Department'] == 'POLICE']
police_salaries
```

```python
police_salaries.isna().sum()
```

```python
# Your code here:
#H0: the salaries this year for the police officers who are salaried are  >= (higher than or equal to) than 86,000/year
#H1: the salaries this year for the police officers who are salaried are  < (lower) than 86,000/year

stat, pval = ttest_1samp(police_salaries['Annual Salary'], 86000)
print('pvalue for the one-tailed test is ', pval/2)
```

```python
police_salaries['Annual Salary'].mean()
```

```python
print('stat is  ', stat)
```

```python
#We do not reject the hypothesis because the mean is greater than 86k and stat is also positive
```

The workers from the department with the most hourly workers have complained that their hourly wage is less than $35/hour. Using a one sample t-test, test this one-sided hypothesis at the 95% confidence level.

```python
Annual_hourly = salaries[salaries['Salary or Hourly'] == 'Hourly'] 
Annual_hourly
```

```python
streets_salaries= Annual_hourly[Annual_hourly['Department'] == 'STREETS & SAN']
streets_salaries
```

```python
# Your code here:

#H0: the salaries for the streets and sanitation department are lower than or equal to $35/h  
#H1: the salaries for the streets and sanitation department are higher than $35/h 

stat, pval = ttest_1samp(streets_salaries['Hourly Rate'], 35)
print('pvalue for the one-tailed test is ', pval/2)
print('stat is  ', stat)
streets_salaries['Hourly Rate'].mean()
```
